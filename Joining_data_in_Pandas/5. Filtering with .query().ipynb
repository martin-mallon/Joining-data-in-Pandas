<h1>Filtering with <code>.query()</code></h1>

<h3><font color='grey'>A convenient way to filter DataFrames</font></h3>

<p>The <code>query</code> method allows DataFrame to be filtered according to the query expression. This is given as a string parameter. The results returned will be the observations where the expression is True.</p>

<ul>
    <li><b><code>df.query('')</code></b> - Allows query criteria to be specified, which works as a convenient filter.</li>
    <br>
    <li><b><code>'> / >= / < / <= / =='</code></b> - Basic logic filters.</li>
    <br>
    <li><b><code>'and / or'</code></b> - Allows layering of additional logic.</li>
</ul>

<br>

```python
# libraries
import pandas as pd
import numpy as np

# import
def csv_to_df(file_path):
    df = pd.read_csv(file_path)
    return df

# data
dev = csv_to_df('Data\data_device.csv')
ttl = csv_to_df('Data\data_totals.csv')
```

```python
ttl_lgn = ttl.query('account_login > 4000')
ttl_lgn.head()
```

```python
ttl_ord = ttl.query('account_login > 4000 and order_placed > 400')
ttl_ord.head()
```

```python
dev_mob = dev.query('device_type == "Mobile"')
dev_mob.head()
```

```python
dev_app = dev.query('device_type == "App Android" or device_type == "App iOS"')
dev_app.head(10)
```
