<h1><font color='black'>Semi & anti joins</font></h1>

<br>

<ul>
    <li><b><code>Semi join</code></b> - Allows primary table to be subsetted by values appearing in joined table.</li>
    <br>
    <li><b><code>Anti join</code></b> - Allows primary table to be subsetted by values <u>not appearing</u> in joined table..</li>
</ul>

<br>

```python
# libraries
import pandas as pd
import numpy as np

# import
def csv_to_df(file_path):
    df = pd.read_csv(file_path)
    return df

# data
jrny = csv_to_df('Data\data_journey.csv')
dev = csv_to_df('Data\data_device.csv')
desk = dev[dev['device_type']=='Desktop']

# jrny
jrny.head()
```

<hr>

<br>

<h2>Semi join</h2>

<h3><font color='grey'>Filtering for values that appear in joined table</font></h3>

<br>

<ul>
    <li><b><code>jd_inner</code></b> - A table of desktop users. The result of an inner join of tables <code>jrny</code> and <code>desk</code>.</li>
    <br>
    <li><b><code>jd_filt</code></b> - A version of the original table <code>jrny</code> that has been subsetted for dektop users by applying subset criteria of <code>jrny['session_id'].isin(jd_inner['session_id']</code>.</li>
</ul>

<br>

```python
# filter criteria
jd_inner = jrny.merge(desk, on='session_id', how='inner')

# subset
jd_filt = jrny[jrny['session_id'].isin(jd_inner['session_id'])]

# show
jd_filt.head()
```

<hr>

<br>

<h2>Anti join</h2>

<h3><font color='grey'>Filtering for values that <u>do not</u> appear in joined table</font></h3>

<br>

<ul>
    <li><b><code>jd_left</code></b> - A left join of tables <code>jrny</code> and <code>desk</code>. Setting <code>indicator=True</code> reveals the <code>_merge</code> field used identify which ids appear in which tables.</li>
    <br>
    <li><b><code>ids</code></b> - A list of <code>session_id</code> that are non-desktop users. <code>jd_left['_merge']</code> is subsetted for <code>left_only</code> ids, resulting in the ids that do not appear in <code>desk</code> and are, therefore, non-desktop users.</li>
    <br>
    <li><b><code>non-desk</code></b> - A version of the original table <code>jrny</code> that has been subsetted for non-dektop users by applying <code>jrny['session_id'].isin(ids)</code>.</li>
</ul>

<br>

```python
# left join
jd_left = jrny.merge(desk, on='session_id', how='left', indicator=True)

# subset for ids in left only
ids = jd_left.loc[jd_left['_merge']=='left_only', 'session_id']

# subset
non_desk = jrny[jrny['session_id'].isin(ids)]
print('Non desktop users: ' + str(non_desk.shape[0]))
non_desk.head()
```
